```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Grayson-choi/Data/master/Dataset/museum1.csv")

is_university = df["시설명"].str.contains('대학교')

df.loc[is_university == True, '분류'] = '대학'
df.loc[is_university == False, '분류'] = '일반'

df
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Grayson-choi/Data/master/Dataset/museum2.csv")

phone_number = df["운영기관전화번호"].str.split(pat="-", n=3, expand=True)
df['지역번호'] = phone_number[0]

region = {
 '02':"서울시",
 '031':"경기도",'032':"경기도", 
 '033':"강원도",
 '041':"충청도",'042':"충청도",'043':"충청도",'044':"충청도",
 "051":"부산시",
 "052":"경상도","053":"경상도","054":"경상도","055":"경상도",
 "061":"전라도","062":"전라도","063":"전라도",
 "064":"제주도",
 "1577":"기타","070":"기타"
}

df["지역번호"] = df["지역번호"].map(region)

df.rename(columns={"지역번호":"지역명"}, inplace = True)

df


```
